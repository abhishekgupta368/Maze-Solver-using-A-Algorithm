```python
class grid_cell():
    def __init__(self,parent_i,parent_j,f,g,h):
        self.parent_i = parent_i
        self.parent_j = parent_j
        self.f = f
        self.g = g
        self.h = h
    def getParent_i(self):
        return self.parent_i
    def getParent_j(self):
        return self.parent_j
    def getF(self):
        return self.f
    def getG(self):
        return self.g
    def getH(self):
        return self.h
```

```python
def isValid(row,col,n,m):
    if(row>=0 and row<n and col>=0 and col<m):
        return True
    else:
        return False

def isUnBlock(grid,row,col):
    if(grid[row][col]==1):
        return True
    else:
        return False
    
def isDestination(pair,dest):
    if(pair[0]==dest[0] and pair[1]==dest[1]):
        return True
    else:
        return False
    
def calValH(row,col,dest):
    val = float(math.sqrt((row-dest[0])*(row-dest[0])+(col-dest[1])*(col-dest[1])))
    return val
```

```python
def aStarSearch(grid,src,dest):
    n = len(grid)
    m = len(grid[0])
    if(isValid(src[0],src[1],n,m)==False):
        print("Source is Invalid")
        return
    if(isValid(dest[0],dest[1],n,m)==False):
        print("Destination is Invalid")
        return
    if(isUnBlock(grid,src[0],src[1])==False):
        print("Source is blocked")
        return
    if( isUnBlock(grid,dest[0],dest[1])==False):
        print("Destination is blocked")
        return
    if( isDestination(src,dest)==True):
        print("Already on Destination")
        return
    
    closedList=[]
    for i in range(n):
        l=[]
        for j in range(m):
            l.append(False)
        closedList.append(l)
    #printData(closedList)
    cellDetail=[]
    i=0
    j=0;
    for i in range(n):
        l=[]
        for j in range(m):
            data= grid_cell(-1,-1,float("inf"),float("inf"),float("inf"))
            l.append(data)
        cellDetail.append(l)
    #printData(cellDetail)
    i,j= src
    cellDetail[i][j].f=0.0
    cellDetail[i][j].g=0.0
    cellDetail[i][j].h=0.0
    cellDetail[i][j].parent_i=i
    cellDetail[i][j].parent_j=j
    heap=[]
    heapq.heappush(heap,(0.0,i,j))
    foundDest=False
    
    while heap:
        #p = openList.pop()
        p= heapq.heappop(heap)
        #print(p)
        i=p[1]
        j=p[2]
        closedList[i][j]=True
        gNew= float("0.0")
        hNew= float("0.0")
        fNew= float("0.0")
        
        mov_x = [0,1,0,-1,0,-1,1,-1]
        mov_y = [0,0,1,0,-1,1,-1,-1]
        
        for x in mov_x:
            for y in mov_y:
                if(isValid(i+x,j+y,n,m)==True):
                    val_src = [i+x,j+y]
                    if(isDestination(val_src,dest)==True):
                        cellDetail[i+x][j+y].parent_i = i
                        cellDetail[i+x][j+y].parent_j = j
                        print("Destination Found")
                        tracePath(cellDetail,dest)
                        foundDest=True
                        return

                    elif(closedList[i+x][j+y]==False and isUnBlock(grid,i+x,j+y)==True):
                        gNew = cellDetail[i][j].g+1.414
                        hNew = calValH(i+x,j+y,dest)
                        fNew = gNew+hNew

                        if(cellDetail[i+x][j+y].f==float("inf") or cellDetail[i+x][j+y].f>fNew):
#                             openList.append((fNew,i+x,j+y))
                            heapq.heappush(heap,(fNew,i+x,j+y))
#                             [openList.append(x) for x in openList if x not in openList]
                            cellDetail[i+x][j+y].f=fNew
                            cellDetail[i+x][j+y].g=gNew
                            cellDetail[i+x][j+y].h=hNew
                            cellDetail[i+x][j+y].parent_i=i
                            cellDetail[i+x][j+y].parent_j=j
                            
#                     openList.sort(key = lambda ky:ky[0])
        
        #print(openList)
    #printData(cellDetail)
    if(foundDest == False):
        print("Failed to find the Destination Cell") 
    return; 
```
